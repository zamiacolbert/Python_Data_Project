```python
location_counts = df['job_country'].value_counts().head(5)
location_counts.sort_index()
plt.bar(location_counts.index, location_counts.values)
plt.title('Number of Jobs per Country')
plt.xlabel('Country')
plt.xticks(rotation = 45, ha = 'right')
plt.ylabel('Number of Postings')
plt.show()
```

```python
salary_by_title = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
plt.barh(salary_by_title.index, salary_by_title.values)
plt.title('Average Salary by Job Title')
plt.xlabel('Average Salarys')
plt.show()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
monthly_avg_salary = df.groupby('job_posted_month')['salary_year_avg'].mean().sort_index()
plt.plot(monthly_avg_salary.index, monthly_avg_salary.values, color='red', label='Average Salary')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job Month')
plt.legend()
plt.show()
```

#PANDAS PLOTTING

```python
job_counts = df['job_title_short'].value_counts()

job_counts
```

```python
#Plotting it with only Matplotlib:
plt.bar(job_counts.index, job_counts)
plt.show()
```

```python
#this is the same chart but using Pandas to plot it instead
job_counts.plot(kind='bar')
plt.show()
```

```python
#We can even customize all the labels just like we did in Matplotlib.

job_counts.plot(kind='bar')
plt.xlabel('')
plt.ylabel('Count of Postings')
plt.title('Job Postings by Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
