```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## reset_index()

```python
df_usa = df[df['job_country'] == 'United States']
```

```python
df_usa.head()
#because we filtered out certain data, the index no longers goes in iteration one by one
```

```python
df_usa.reset_index(inplace=True)
df_usa
#technically could go through and drop the job index if we want, but luke would argue to keep it. 
#will be useful in the future when murging dataframes by index
```

## set_index()

set the dataframe index using existing columns

```python
df_usa.set_index('job index', inplace=True)
```

```python
df_usa
#job index back but other index dropped
```

## .sort_index()

returns a new dataframe sorted by label

```python
#useful for pivoting or grouping data
```

```python
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median','min','max'])
median_pivot
```

```python
#can sort median with sort_values
#it actually has 2 level of column titles median then salary_year_avg, so it need yo be provided as a tuple
median_pivot.sort_values(by=('median','salary_year_avg'), ascending =False)
```

```python
#simimlary, we can use sort_index
median_pivot.sort_index() 
#the index (job_title_short in this case) is sorted alphabetically
```

```python
median_pivot.sort_index(inplace=True) #to modify current dataframe
```

```python
median_pivot
```

```python
#could also use sort_values to sort index alphbetically (in desc order now bc asc=false)
#sort_index can be useful for when the index doesnt' have a name
median_pivot.sort_values(by='job_title_short', ascending =False)
```
